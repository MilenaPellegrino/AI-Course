```python
datos_na = datos.dropna()
```

```python
datos_na.info()
```

```python
#Convertirmos el texto de la proximidad al oceano a numerico
datos_na["ocean_proximity"]
```

```python
datos_na["ocean_proximity"].value_counts()
```

```python
# Dummies / One-hot encoding
# NEAR BAY    INLAND    NEAR  OCEAN
# 1             0         0    0
dummies = pd.get_dummies(datos_na["ocean_proximity"])
```

```python
datos_na = datos_na.join(dummies)
```

```python
datos = datos_na.drop(["ocean_proximity"], axis=1)
```

```python
#Analisis y correlaciones
datos.corr()
```
