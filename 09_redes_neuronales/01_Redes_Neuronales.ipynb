```python
import pandas as pd
datos = pd.read_csv("celsius.csv")
```

```python
datos.head()
```

```python
datos["celsius"].values
```

```python
x = datos["celsius"].values
y = datos["fahrenheit"].values
```

```python
import tensorflow as tf

modelo = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)), # La entrada es 1 sola, los grados celsius
    tf.keras.layers.Dense(units=1), # La salida tambien es 1, los grados fahrenheit
])

modelo.compile(
    optimizer = tf.keras.optimizers.Adam(.1),
    loss = "mean_squared_error"
)
```

```python
entrenamiento = modelo.fit(x, y, epochs=2000)
```

```python
import matplotlib.pyplot as plt

plt.xlabel("# Epoca")
plt.ylabel(" Magnitud de la perdida")
plt.plot(entrenamiento.history["loss"])
```

```python
c= 123
prediccion = modelo.predict([c])
print(f"La prediccion de {c} grados celsius son {prediccion[0]}")
```
